4. Generate a dataset with categories and numerical values. Visualize the distribution of a numerical
variable across different categories.

```python


# Generate sample data
np.random.seed(42)
categories = ['A', 'B', 'C']
num_values = 100
data = {'Category': np.random.choice(categories, size=num_values),
        'Value': np.random.randn(num_values)}
df = pd.DataFrame(data)

sns.boxplot(x='Category', y='Value', data=df)  # Box plot

# Add labels and title
plt.xlabel("Category")
plt.ylabel("Value")
plt.title("Distribution of Numerical Variable across Categories")

# Show plot
plt.show()
```

5. Generate a synthetic dataset with correlated features. Visualize the correlation matrix of a dataset using a
heatmap.

```python
np.random.seed(42)
num_samples = 100
num_features = 4

# Create a random dataset
data = np.random.rand(num_samples, num_features)

# Introduce correlation
data[:, 1] = data[:, 0] + np.random.rand(num_samples) * 0.1  # Correlate feature 1 with feature 0

# Create a Pandas DataFrame
df = pd.DataFrame(data, columns=['Feature_' + str(i) for i in range(num_features)])

# Calculate correlation matrix
correlation_matrix = df.corr()

# Visualize correlation matrix using Seaborn heatmap
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm')

# Add title
plt.title("Correlation Matrix Heatmap")

# Show plot
plt.show()
```

4. Using the given x and y data, generate a 3D surface plot to visualize the function


```python
import numpy as np
import plotly.graph_objects as go
x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)
x, y = np.meshgrid(x, y)
z = np.sin(np.sqrt(x**2 + y**2))
data = {
    'X': x.flatten(),
    'Y': y.flatten(),
    'Z': z.flatten()
}
df = pd.DataFrame(data)



# Create 3D surface plot
fig = go.Figure(data=[go.Surface(x=x, y=y, z=z, colorscale='Viridis', opacity=0.8)])

# Set plot title and axis labels
fig.update_layout(title='3D Surface Plot of sin(x^2 + y^2)', scene=dict(xaxis_title='X', yaxis_title='Y', zaxis_title='Z'))

# Show the plot
fig.show()
```

##(Use Bokeh for the visualization of the given questions)

1.Create a Bokeh plot displaying a sine wave. Set x-values from 0 to 10 and y-values as the sine of x

```python
import bokeh
from bokeh.io import show, output_notebook
from bokeh.plotting import figure

x=np.linspace(0,10,100)
y=np.sin(x)
p=figure(title="sine wave",x_axis_label="x",y_axis_label="y")
p.line(x,y,legend_label="sine",line_width=2)
output_notebook()
show(p)

```

2.Create a Bokeh scatter plot using randomly generated x and y values. Use different sizes and colors for the
markers based on the 'sizes' and 'colors' columns.

```python


N = 100
x = np.random.random(size=N) * 10  # Random x values between 0 and 10
y = np.random.random(size=N) * 10  # Random y values between 0 and 10
sizes = np.random.randint(5, 20, size=N)  # Random sizes between 5 and 20
colors = np.random.choice(["red", "blue", "green"], size=N)  # Random colors

# Create a Bokeh plot
p = figure(title="Scatter Plot", x_axis_label="x", y_axis_label="y")

# Add a scatter glyph with varying sizes and colors
p.scatter(x, y, size=sizes, color=colors, alpha=0.6)  # alpha for transparency

# Show the plot
show(p)
```

3. Generate a Bokeh bar chart representing the counts of different fruits using the following dataset

```python
fruits = ['Apples', 'Oranges', 'Bananas', 'Pears']
counts = [20, 25, 30, 35]
p=figure(x_range=fruits,title="fruits counts",x_axis_label="fruits",y_axis_label="counts")
p.vbar(x=fruits,top=counts,width=0.5)
show(p)
```

4. Create a Bokeh histogram to visualize the distribution of the given data.

```python
data_hist = np.random.randn(1000)
hist, edges = np.histogram(data_hist, bins=30)

p=figure(title="histogram",x_axis_label="value",y_axis_label="frequency")
p.quad(top=hist,bottom=0,left=edges[:-1],right=edges[1:],fill_color="skyblue",line_color="black")
show(p)
```

5. Create a Bokeh heatmap using the provided dataset.

```python


# Data for the heatmap
data_heatmap = np.random.rand(10, 10)
x = np.linspace(0, 1, 10)
y = np.linspace(0, 1, 10)
xx, yy = np.meshgrid(x, y)

# Create a Bokeh plot
p = figure(title="Heatmap", x_axis_label="x", y_axis_label="y",
           x_range=(0, 1), y_range=(0, 1),  # Set x and y ranges
           tools="hover", tooltips=[("x", "$x"), ("y", "$y"), ("value", "@image")])  # Enable hover tooltips

# Add an image glyph to represent the heatmap
p.image(image=[data_heatmap], x=0, y=0, dw=1, dh=1, palette="Viridis256")

# Show the plot
show(p)
```
